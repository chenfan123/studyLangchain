##### 模型调用中config参数
如使用 invoke(), ainvoke(), stream(),batch()等方法时，允许在调用模型时，动态地配置和控制模型的行为，而无需在初始化时就固定所有参数，这为应用带来了极大的灵活性和可维护性。

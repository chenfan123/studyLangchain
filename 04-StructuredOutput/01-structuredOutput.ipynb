# 结构化输出
要求模型最终返回一个符合预定义结构的数据对象，例如：固定字段的JSON、Pydantic模型、TypedDict等
目标：自然语言回答 变成 程序可以稳定消费的数据
## 结构化输出模式
1. Pydantic （字段校验、描述、嵌套结构）
2. TypedDict（轻量类型约束）
3. JSON Schema （与前后端/跨端接口最通用）
4. dataclass 

模型对象可以调用with_structured_output() 方法，绑定输出模式。

只有 `Pydantic` 返回的是`Schema类实例`，其余三种方式返回的都是 `字典` ；也只有 Pydantic 在类型不匹配时会抛出异常。
